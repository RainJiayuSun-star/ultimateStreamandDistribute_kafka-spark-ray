```python
import pandas as pd
import os
import sys
```

```python
# data directory
data_dir = '../src/data'
data_dir_historical = os.path.join(data_dir, "historical")
print(os.listdir(data_dir_historical))

data_dir_list = os.listdir(data_dir_historical)
```

```python
# read in files
"""
# example
madison2025_path = os.path.join(data_dir_historical, os.listdir(data_dir_historical)[1])
madison2025_df = pd.read_csv(f'{madison2025_path}')
print(madison2025_df.info())
madison2025_df.head(10)
"""

# load the 2016-2025 file (contains all weathers data from the airport station)
madison2016_2025_path = os.path.join(data_dir_historical, os.listdir(data_dir_historical)[1])
madison2016_2025_df = pd.read_csv(f'{madison2016_2025_path}')
print(madison2016_2025_df.info())

# load the 1996-2015 file (contains all weathers data from the airport station)
madison1996_2015_path = os.path.join(data_dir_historical, os.listdir(data_dir_historical)[0])
madison1996_2015_df = pd.read_csv(f'{madison1996_2015_path}')
print(madison2016_2025_df.info())
madison2016_2025_df.head(10)
```

```python
# combine the data
# use a list to stack them vertically
combined_df = pd.concat([madison1996_2015_df, madison2016_2025_df], ignore_index=True)

# 2. Convert the date column to datetime objects
# Replace 'DATE' with the actual name of your date column (e.g., 'timestamp' or 'date')
combined_df['DATE'] = pd.to_datetime(combined_df['DATE'])

# 3. Sort by the date column
combined_df = combined_df.sort_values(by='DATE').reset_index(drop=True)

# Verify the results
print(combined_df.head())
print(combined_df.info())
combined_df.head(20)
```

```python
# clean the combined data (theoretically computationally more expensite, this cost factor is ignored here.)
# We only want the ones related to Temperature, Pressure, Wind, and Precipication.
# Temperature: TMAX, TMIN
# Pressure: NA
# Wind: AWND, Average Wind Speed
# Precipication: PRCP
# Keep STATION, NAME, LATITUDE, LONGITUDE, ELEVATION, DATE, and the weather factors
columns_to_keep = [
    'STATION', 'NAME', 'LATITUDE', 'LONGITUDE', 'ELEVATION', 'DATE', # Metadata
    'TMAX', 'TMIN', 'TAVG',                                          # Temperature
    'AWND',                                                         # Wind
    'PRCP'                                                          # Precipitation
]
df_cleaned = combined_df[combined_df.columns.intersection(columns_to_keep)]
df_cleaned.head(10)
```

```python
# save the combined and cleaned dataframe
combined_df_name = "madison1996_2025_combined_cleaned.csv"
try:
    combined_df_path = os.path.join(data_dir_historical, f"{combined_df_name}")
    df_cleaned.to_csv(f'{combined_df_path}', index=False)
    print(f"Successfully saved combined data to {combined_df_path}")
except PermissionError:
    print(f"Error: Could not save file. Please check if '{combined_df_name}' is open in Excel or another program.")
except FileNotFoundError:
    print(f"Error: The director '{data_dir_historical}' does not exist.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
